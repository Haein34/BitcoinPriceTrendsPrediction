```python
from sklearn.model_selection import train_test_split

data = df[['Date','Volume_total','sentiment_index_total','High','Low','Vol',
                        'RSI','SMA_5','SMA_20','SMA_60','EMA','MACD','signal','Stochastic RSI_fastk',
                        'Stochastic RSI_fastd','Stochastic Oscillator Index_slowk','Stochastic Oscillator Index_slowd',
                        ]]
target = df['Fluctuations']

x_train, x_valid, y_train, y_valid = train_test_split(data, target, test_size=0.2, shuffle=False, random_state=34)
```
